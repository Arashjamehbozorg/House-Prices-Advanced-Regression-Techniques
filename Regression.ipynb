```python
import pandas as pd
import numpy as np
```

```python
"""
LotFrontage: Linear feet of street connected to property

LotArea: Lot size in square feet

"""

train_data= pd.read_csv("train.csv", index_col=0)
test_data=pd.read_csv("test.csv", index_col=0)
target = train_data["SalePrice"]
numerical_train_data= train_data.select_dtypes(include=[np.number]).columns.tolist()
numerical_train_data
correlation_train_data = train_data[numerical_train_data]
correlation_train_data
```

```python
train_data
```

```python
numerical_train_data_copy = numerical_train_data.copy()
train_df=train_data[numerical_train_data_copy]
target=train_data["SalePrice"]

if "SalePrice" in numerical_train_data_copy:
  numerical_train_data_copy.remove("SalePrice")

test_df=test_data[numerical_train_data_copy]
test_df=test_df.fillna(test_df.median())
train_df = train_df.fillna(train_df.median())

train_df.isnull().sum().sum()
```

```python
train_df
```

```python
# calculating correlation
corr = correlation_train_data.corr()["SalePrice"].sort_values(ascending=False)
high_corr_features = corr[abs(corr) > 0.1].index.tolist()
high_corr_features.remove("SalePrice")
train_df_highcorr = train_df[high_corr_features]
test_df_highcorr = test_df[high_corr_features]
train_df_highcorr
train_df_highcorr_total_bath = train_df_highcorr.copy()
train_df_highcorr_total_bath['TotalBathrooms'] = (train_df['FullBath'] + train_df['BsmtFullBath'] +
                             (train_df['HalfBath'] + train_df['BsmtHalfBath']) * 0.5)
train_df_highcorr_total_bath
len(train_df_highcorr_total_bath.columns)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(train_df_highcorr_total_bath, target)
```

```python
test_df_highcorr_total_bath = test_df_highcorr.copy()
test_df_highcorr_total_bath['TotalBathrooms'] = (test_df['FullBath'] + test_df['BsmtFullBath'] +
                             (test_df['HalfBath'] + test_df['BsmtHalfBath']) * 0.5)

result = model.predict(test_df_highcorr_total_bath)
```

```python
result
```

```python
# Create a DataFrame with Id from test_data and the prediction results
submission_df = pd.DataFrame({'Id': test_data.index, 'SalePrice': result})

# Display the first few rows of the submission DataFrame
display(submission_df.head())

# Save the DataFrame to a CSV file
submission_df.to_csv('submission_v9.csv', index=False)
```
